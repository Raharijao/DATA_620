```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

#edges
edges = [
    ("David", "Sarah"), ("David", "John"), ("David", "Patrick"), ("David", "Madeline"),
    ("Sarah", "John"), ("Sarah", "Patrick"), ("Sarah", "Madeline"),
    ("John", "Patrick"), ("John", "Madeline"),
    ("Patrick", "Sydney"),
    ("Madeline", "Patrick"), ("Madeline", "Cody"),
    ("Sydney", "Cody"), ("Sydney", "Laura"),
     ("Cody", "Kyle"),
    ("Laura", "Kyle"), ("Laura", "Chris"),
    ("Kyle", "Chris")
]




G.add_edges_from(edges)
nx.draw(G, with_labels=True, node_color="lightblue", node_size=600, font_size=10)
plt.show()
```
