```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#loading Zachary karate club network example
G = nx.karate_club_graph()
```

```python
#basic info
num_node = G.number_of_nodes()
num_edge = G.number_of_edges()

print(num_node)
print(num_edge)
```

```python
#graph diameter
diameter = nx.diameter(G)
print(diameter)
```

```python
bet_centrality = nx.betweenness_centrality(G)

df_bet = pd.DataFrame(list(bet_centrality.items()), columns=["Node", "betweenness"])
df_bet = df_bet.sort_values(by="betweenness", ascending=False).reset_index(drop=True)
df_bet
```

```python
nx.draw(G, with_labels=True, node_size=200)
plt.show()
```

```python
nx.write_gexf(G, "karate_subset.gexf")
```
